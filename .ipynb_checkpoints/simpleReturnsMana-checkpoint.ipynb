```python
import numpy as np
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
manaUsd = wb.DataReader('MANA-USD', data_source='yahoo', start='2020-2-1',end='2021-1-1')
```

```python
manaUsd.head(20) 
```

```python
manaUsd.tail(20)
```

```python
manaUsd['simple_return'] = (manaUsd['Adj Close'] / manaUsd['Adj Close'].shift(1)) -1 
print(manaUsd['simple_return'])
```

```python
manaUsd['simple_return'].plot(figsize=(10,5))
plt.show()
```

```python
avg_returns_d = manaUsd['simple_return'].mean()
avg_returns_d
```

```python
avg_returns_a = manaUsd['simple_return'].mean() * 250
avg_returns_a 
```

```python
print(str(round(avg_returns_a, 5) * 100) + ' %')
```
